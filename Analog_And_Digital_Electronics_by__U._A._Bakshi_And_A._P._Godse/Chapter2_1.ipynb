## Example 2.4 Page No. 2-11 

```python
import math 

print("It is necessary to analyze each network to determine the critical frequency of the amplifier");


print("(a) Input RC network ")
fc1 = 1/(2*(math.pi)*(680+1031.7)*(0.1*pow(10,-6)) )
print "fc(input) (in Hz) = 1/2*pi*[RS+(R1||R2||hie)]C1 =",round(fc1,1),"Hz"


print("\n\n(b) Output RC network ")
fc2 = 1/(2* (math.pi) *((2.2+10)*pow(10,3)) *(0.1*pow(10,-6)))
print "fc(output) (in Hz) = 1/2*pi*(RC+RL)*C2 =",round(fc2,2),"Hz"


print("\n\n(c) Bypass RC network")
rth = ((68*22*0.680) /((22*0.680) +(68*0.680) +(68*22)))*pow(10,3)
print "Rth (in ohm) = R1||R2||RS =",rth
fc3 = 1/(2* (math.pi) *17.23*10*pow(10,-6))
print " fc (bypass) (in Hz) = 1/2*pi*[ (Rth+hie/beta)||RE]*CE = ",round(fc3,1),"Hz"

print("\n\nWe have calculated all the three critical frequencies: ")
print("(a) fc (input) = 929.8 Hz")
print("(b) fc (output) = 130.45 Hz")
print("(c) fc (bypass) = 923.7 Hz")
```

## Example 2.6 Page No. 2-17 

```python
import math

print("Before calculating critical frequencies it is necessary calculate mid frequency gain of the  given circuit. This is required to calculate Cin(miller) and Cout(miller)")
print("\nAv= hfe*Ro/Ri")
print("where Ri = hie||R1||R2")
print("and Ro = RC||RL")

av = (-100*1.8)/1.032
print "\nTherefore, Av = hfe(RC||RL)/hie||R1||R2=",round(av,1)
print("Negative sign indicates 180 degree shift between input and output")

cin =(4*(174.4+1))*pow(10,-3)
print "\nCin(miller)(in nF) = C_bc*(Av+1) =",round(cin,4),"nF"

cout =(4*175.4) /(174.4)
print "\nCout(miller)(in pF) = C_bc*(Av+1)/Av =",round(cout,1),"pF"
print("We know analyze input and output network for critical frequency")

fci =(1/(2* (math.pi) *410*0.7216*pow(10,-9)) ) *pow(10,-3)
print "\nfc(input)(in kHz) = 1/2*pi*(Rs||R1||R2||hie)*(C_be+Cin(miller)) =",round(fci,3),"Hz"

fco =(1/(2* (math.pi) *((22*pow(10,6)) /(12.2*pow(10,3)) ) *(4*pow(10,-12)) ))*pow(10,-6)
print "\nfc(output)(in MHz) = 1/2*pi*(RC||RL)*Cout(miller) =",round(fco,1),"MHz"

print("\nWe have calculated both the critical frequencies")
print("(a) fc(input) = 537.947 kHz")
print("(b) fc(output) = 22.1 MHz")
```
