```python
import math

print("From the given information we can write,")
print("A = −16∗10ˆ6/j∗omegaandbeta=10ˆ3/[2∗10ˆ3+j∗omega]ˆ2")
print("To verify the Barkhausen condition means to verify whether|A∗beta|=1 at a frequency for which A∗beta=0 degree.Let us express,A∗beta in its rectangluar form.")
print("A∗beta=−16∗10ˆ6∗10ˆ3/j∗omega∗[2∗10ˆ3+j∗omega]ˆ2=−16∗10ˆ9/j∗omega∗[4∗10ˆ6+4∗10ˆ3∗j∗omega+(j∗omega)ˆ2]")
print("=−16∗10ˆ9/j∗omega∗[4∗10ˆ6+4∗10ˆ3∗j∗omega−omegaˆ2] as j∗2=−1")
print("=−16∗10ˆ9/4∗10ˆ6∗j∗omega+4∗10ˆ3∗jˆ2∗omegaˆ2−j∗omegaˆ3]")
print("=−16∗10ˆ9/j∗omega∗[4∗10ˆ6−omegaˆ2]−[omegaˆ2∗4∗10ˆ3]")
print("Rationalising the denominator function we get,")
print("A∗beta=−16∗10ˆ9[−omegaˆ2∗4∗10ˆ3−j∗omega∗[4∗10ˆ6−omegaˆ2]]/[−[omegaˆ2∗4∗10ˆ3]−j∗omega∗[4∗10ˆ6−omegaˆ2]]∗[−omegaˆ2∗4∗10ˆ3−j∗omega∗[4∗10ˆ6−omegaˆ2]]")
print("Using (a−b)(a+b)=aˆ2−bˆ2 in the denominator,")
print("A∗beta=16∗10ˆ9[omegaˆ2∗4∗10ˆ3+j∗omega∗[4∗10ˆ6−omegaˆ2]]/[−omegaˆ2∗4∗10ˆ3]ˆ2−[j∗omega∗[4∗10ˆ6−omegaˆ2]ˆ2")
print("A∗beta=16∗10ˆ9[omegaˆ2∗4∗10ˆ3+j∗omega∗[4∗10ˆ6−omegaˆ2]]/16∗10ˆ6∗omegaˆ4+omegaˆ2(4∗10ˆ6−omegaˆ2)ˆ2")
print("Now to have A∗beta=0degree,the imaginary part of A∗beta must be zero.This is possible when,")
print("Therefore,omega∗(4∗10ˆ6−omegaˆ2)=0")
print("Therefore,omega=0 or 4∗10ˆ6−omegaˆ2=0")
print("Therefore,omegaˆ2=4∗10ˆ6 Neglecting zero value of frequency")
print("Therefore,omega=2∗10ˆ3rad/sec")
print("At this frequency |A∗beta| can be obtained as,")
print("|A∗beta| = 16∗10ˆ9[4∗10ˆ3∗omegaˆ2]/16∗10ˆ6∗omegaˆ4+omegaˆ2[4∗10ˆ6−omegaˆ2]ˆ2atomega=2∗10ˆ3")
ab=(2.56*pow(10,20))/(2.56*pow(10,20))
print "|A∗beta|=",round(ab,1)
print("Therefore,At omega=2∗10ˆ3rad/sec,A∗beta = 0 degree as imaginary part is zero while |A∗beta|=1.Thus Barkhausen Criterion is satisfied.")
print("The frequency at which circuit will oscillate is the value of omega for which|A∗beta| = 1 and A∗beta=0 degree at the same time")
print("i.e.omega=2∗10ˆ3rad/sec")
print("But omega=2∗pi∗f")
f=(2.0*pow(10,3))/(2.0*math.pi)
print "Therefore,f(in Hz)=omega/2pi=",round(f,3),"Hz"
```
