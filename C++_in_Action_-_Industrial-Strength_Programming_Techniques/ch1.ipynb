### EXAMPLE 1.3 Page no : 4

```python
'''
Naming accessors and modifiers
'''
class Point:
    def __init__(self,a,b):
        self.x = a
        self.y = b
    def x1(self,a):
        self.x = a

p = Point(0,0) # a point in a 2-dimensional space
p.x1(1); # set X-coordinate
print  p.x  # prints X-coordinate, "1"
```
